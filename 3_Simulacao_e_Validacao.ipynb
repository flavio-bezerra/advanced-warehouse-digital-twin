# 3. Simulação e Validação (Validacao)

**Objetivo:** Validar o motor de simulação TSP.


```python
import sys
import os
import pandas as pd
sys.path.append(os.path.abspath('.'))

from src import simulation_engine

# Carregar dados
try:
    df_orders = pd.read_csv('data/pedidos_backlog.csv')
    # Tenta carregar o otimizado, senão o inicial
    if os.path.exists('data/mapa_alocacao_otimizado.csv'):
        df_alloc = pd.read_csv('data/mapa_alocacao_otimizado.csv')
        print("Usando alocação otimizada.")
    else:
        df_alloc = pd.read_csv('data/mapa_alocacao_inicial.csv')
        print("Usando alocação inicial.")
        
    df_layout = pd.read_csv('data/layout_fisico.csv')
except FileNotFoundError:
    print("Arquivos de dados não encontrados.")

# Rodar Simulação (Amostra pequena para teste rápido)
df_kpis = simulation_engine.run_simulation(
    df_orders, df_alloc, df_layout, 
    num_orders_to_sim=20, 
    forklift_speed=1.5, 
    num_active_docks=1
)
df_kpis.to_csv('data/kpis_simulacao.csv', index=False)

print("Resultados:")
print(df_kpis.head())
```
